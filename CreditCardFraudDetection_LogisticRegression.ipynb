# Credit Card Fraud Detection - Logistic Regression

## Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## Generate a smaller dataset for better processing

```python
# loading the dataset to a Pandas DataFrame
df = pd.read_csv('/content/sample_data/creditcard.csv')
```

```python
# Filter fraud cases (Class = 1)
fraud_cases = df[df['Class'] == 1]
```

```python
fraud_cases
```

```python
# Filter non-fraud cases (Class = 0) and randomly sample 5000 rows
non_fraud_cases = df[df['Class'] == 0].sample(n=5000, random_state=42)
```

```python
non_fraud_cases
```

```python
# Combine the two datasets
smaller_dataset = pd.concat([fraud_cases, non_fraud_cases])
```

```python
# Shuffle the dataset
credit_card_data = smaller_dataset.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
credit_card_data
```

## Dataset info

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# dataset informations
credit_card_data.info()
```

```python
# checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit transactions & fraudulent transactions
credit_card_data['Class'].value_counts()
```

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)
```

```python
# Concatenating two DataFrames
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

## Splitting the data into Features & Targets



```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

## Split the data into Training data & Testing Data



```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Model Training - Logistic Regression

```python
model = LogisticRegression(solver='liblinear')
```

```python

# Fit Logistic Regression with scaled data
model.fit(X_train, Y_train)
```

## Model Evaluation

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
